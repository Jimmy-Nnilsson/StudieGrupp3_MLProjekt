```python
import matplotlib.pyplot as plt
import pathlib
from PIL import Image
import numpy as np
import cv2
```

## Väljer en random brain scan
### Plottar original och en förminskad version

```python
file_path = pathlib.Path(r'C:\Users\AndreasP\Desktop\work\BeyondExperience\images\Y13.jpg')

pic = cv2.imread(str(file_path))
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)      # måste vända på kanalerna!
small_pic = cv2.resize(pic,(50,50))

fig, ax = plt.subplots(1,2, figsize = (12, 6))
fig.suptitle("Original vs resized", fontsize =22)
ax[0].imshow(pic)
ax[0].set_title(pic.shape)
ax[1].imshow(small_pic)
ax[1].set_title(small_pic.shape)
plt.show()
```

## Plottar olika resize-metoder

```python
type_method =       [cv2.INTER_NEAREST, cv2.INTER_LINEAR , cv2.INTER_CUBIC , cv2.INTER_AREA , cv2.INTER_LANCZOS4 ,cv2.INTER_LINEAR_EXACT , cv2.INTER_NEAREST_EXACT]
type_method_str =   ["cv2.INTER_NEAREST", "cv2.INTER_LINEAR" , "cv2.INTER_CUBIC" , "cv2.INTER_AREA" , "cv2.INTER_LANCZOS4" ,"cv2.INTER_LINEAR_EXACT" , "cv2.INTER_NEAREST_EXACT"]
row = 2
col = 4
fig, ax = plt.subplots(row, col, figsize = (20, 10))
i = 0
for r in range(row):
    for c in range(col):
        if i< len(type_method):
            resiz_pic = cv2.resize(small_pic, (250, 250), interpolation = type_method[i] )
            ax[r][c].imshow(resiz_pic)
            ax[r][c].set_title(type_method_str[i])
            i+=1
plt.show()
```
